```python
import numpy as np     # use uncertainty package to handle error propagation 
from uncertainties import ufloat
from uncertainties.umath import *
import statistics
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
run1 = [20.08,20.045,20.04,20.04,20.035,20.03,20.03,20.03,20.025,20.025,20.02,20.02,20.35,20.99,21.74,21.97,22.11,22.23,22.28,22.29,22.33,22.33,22.34,22.34,22.345,22.345,22.345,22.35,22.345,22.345,22.345,22.34,22.34,22.335]
run2 = [18.920,18.920,18.920,18.920,19.150,20.15,21.30,21.8,22.05,22.17,22.24,22.28,22.30,22.32,22.33,22.335,22.34,22.34,22.345,22.35,22.35,22.35]
time1 = list(range(0,991,30))
time2 = list(range(0,632,30))
```

```python
def linearline(x,m,b):
    y = []
    for i in x:
        y.append(i*m+b)
    return y
```

```python
params11, _ = curve_fit(linearline,time1[1:12],run1[1:12])
params12, _ = curve_fit(linearline,time1[27:],run1[27:])
print(params11,params12)
```

```python
plt.figure(figsize=(20,10))
plt.plot(time1[:2],run1[:2],color="green",linewidth=3.9,label="System reaching thermal equilibrium")
plt.plot(time1[1:12],run1[1:12],color="blue",linewidth=3.9,label='System at equilibrium, before combustion')
plt.plot(time1[11:28],run1[11:28],color="red",linewidth=3.9,label='After combustion, not at equilibrium')
plt.plot(time1[27:],run1[27:],color="purple",linewidth=3.9,label='Equilibrium, after combustion')
plt.plot(list(range(1000)),linearline(list(range(1000)),*params11),label='Projection of heat being lost by the system')
plt.plot(list(range(1000)),linearline(list(range(1000)),*params12),label='Projection of heat being lost by the system')
plt.legend(loc='best', bbox_to_anchor=(0.5, 0.1, 0.5, 0.5),fontsize='14')
plt.fill_between(time1[11:15],linearline(list(range(330,421,30)),*params11),run1[11:15], color='grey', alpha='0.5')
plt.fill_between(time1[14:28],run1[14:28],linearline(list(range(420,811,30)),*params12), color='grey', alpha='0.5')
plt.xlabel('Time (seconds)',fontsize='16')
plt.ylabel('Temperature (Celsius)',fontsize='16')
plt.title('Temperature vs. Time',fontsize='22')
plt.show()
```

```python
A1 = (sum(run1[11:15])+sum(run1[12:14]))*15 - (sum(linearline([330,420],*params1))*(420-330)/2)

A2 = (sum(linearline([420,810],*params2))*(810-420)/2) - (sum(run1[14:28])+sum(run1[15:27]))*15
print(A1,A2)
#the temperature at t=420s is the closest to these areas being equal
```

```python
y1 = linearline([330,420,810],*params1)
y2 = linearline([330,420,810],*params2)
deltaT = (y2[2]-y1[0]-params1[0]*(420-330)-params2[0]*(810-420))
print(deltaT)
```

```python
#find change in heat
Eba = 26.41*ufloat(0.9063,.0002)
Ew = 6.68*(ufloat(.0541,.0002)-ufloat(.0373,.0002))
E = Eba + Ew #in kJ
print(E)
heatcapacity = E/deltaT
print(E/deltaT)
```

```python
plt.figure(figsize=(20,10))
plt.plot(time2[:4],run2[:4],linewidth=3.3,color="blue",label='Time before combustion')
plt.plot(time2[3:20],run2[3:20],linewidth=3.3,color="red",label='Time after combustion, before equilibrium')
plt.plot(time2[-3:],run2[-3:],linewidth=3.3,label='Time after the system reaches equilibrium')
plt.legend(loc='best', bbox_to_anchor=(0.5, 0.1, 0.5, 0.5),fontsize='14')
plt.xlabel('Time (seconds)',fontsize='16')
plt.ylabel('Temperature (Celsius)',fontsize='16')
plt.title('Temperature vs. Time',fontsize='22')
plt.show()
```
